```python
agent_query = '''SELECT
        DATE(sales_agent.created_at) AS created_at,
        sales_agent.type_of_agent,
        sales_agent.email,
        CONCAT(sales_agent.first_name, ' ', sales_agent.last_name) AS agent_name,
        sales_agent.phone_number,
        sales_agent.agent_code
    FROM 
        life_insurance_salesagent AS sales_agent
    WHERE 
        sales_agent.type_of_agent = 'LIBERTY_LIFE_AGENT';
    '''
agents_details  = life_connection(agent_query)
```

```python
sales_query = '''
SELECT 
    DATE(created_at) AS date,
    plan_type,
    insurance_duration,
    amount,
    status,
    channel,
    agent_phone_number
FROM 
    life_insurance_healthplan 
WHERE 
    status = 'SUCCESS'
AND 
    channel != 'SEEDS_BORROWER_CHANNEL'
AND 
    agent_phone_number IS NOT NULL
'''
sales_details = life_connection(sales_query)
```

```python
sales_per_agent = sales_per_agent = sales_details.groupby('agent_phone_number').agg({'amount': 'sum'})
```

```python
agents_sales = pd.merge(sales_per_agent, agents_details, how = "left", left_on = "agent_phone_number", right_on = "phone_number")
```
